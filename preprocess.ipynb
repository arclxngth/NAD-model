```python
import pandas as pd
from sklearn import preprocessing
rawDatas = pd.read_table("./datasets/datasets.txt", delimiter=",")

labels = rawDatas['attack']

del rawDatas['num_outbound_cmds']
del rawDatas['attack']
```

```python
# normalize data
numCols = rawDatas.select_dtypes(exclude="object").columns
min_max_scaler = preprocessing.MinMaxScaler()
rawDatas[numCols] = min_max_scaler.fit_transform(rawDatas[numCols])

# convert string data to one-hot
rawDatas = pd.get_dummies(rawDatas)
rawDatas['is_anomaly'] = labels
# len(xTrain[numCols].columns)
rawDatas.head()
```

```python
rawDatas.iloc[:125973].to_csv("./datasets/pd-train", index=False)
rawDatas.iloc[125973:].to_csv("./datasets/pd-test", index=False)
```
